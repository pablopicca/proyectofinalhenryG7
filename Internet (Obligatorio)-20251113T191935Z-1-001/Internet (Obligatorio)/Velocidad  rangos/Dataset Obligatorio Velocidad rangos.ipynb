```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Importar archivos con miles y decimales correctos
df_prov = pd.read_csv(
    "Internet Accesos Velocidad Rango Provincias.csv",
    thousands=".",
    decimal=","
)

df_tot = pd.read_csv(
    "Internet Accesos Velocidad Rango Totales.csv",
    thousands=".",
    decimal=","
)
```

PASO 8 – Asegurar que las columnas numéricas clave sean numéricas

```python
# Forzamos tipo numérico en 'Accesos' si existe
for df in (df_prov, df_tot):
    if "Accesos" in df.columns:
        df["Accesos"] = pd.to_numeric(df["Accesos"], errors="coerce")

print("=== dtypes PROVINCIAS (tras limpieza ligera) ===")
print(df_prov.dtypes)

print("\n=== dtypes TOTALES (tras limpieza ligera) ===")
print(df_tot.dtypes)
```

PASO 9 – Detección de outliers (IQR) y listado (solo valores altos)

```python
def listar_outliers_superiores(df, nombre_df):
    print(f"\n\n===== OUTLIERS (IQR) – {nombre_df} =====")
    
    numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
    
    for col in numeric_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lim_sup = q3 + 15 * iqr
        
        outliers = df[df[col] > lim_sup].sort_values(col, ascending=False)
        
        print(f"\nColumna: {col}")
        print(f"Límite superior: {lim_sup:,.2f}")
        print(f"Cantidad de outliers superiores: {len(outliers)}")
        display(outliers.head(10))  # top 10 más grandes

listar_outliers_superiores(df_prov, "PROVINCIAS")
listar_outliers_superiores(df_tot, "TOTALES")
```

PASO 10 – Distribución por rangos de velocidad (si existe columna de rango)

```python
def distribucion_por_rango(df, nombre_df):
    # Buscamos una columna que tenga "rango" en el nombre
    cols_rango = [c for c in df.columns if "rango" in c.lower()]
    if not cols_rango:
        print(f"\n{nombre_df}: no se encontró una columna de 'rango' de velocidad.")
        return
    
    col_rango = cols_rango[0]
    print(f"\n{nombre_df}: usando columna de rango -> {col_rango}")

    if "Accesos" not in df.columns:
        print(f"{nombre_df}: no existe columna 'Accesos' para agrupar.")
        return
    
    acc_por_rango = (
        df.groupby(col_rango)["Accesos"]
          .sum()
          .sort_values(ascending=False)
    )
    
    print(f"\n=== Accesos por rango de velocidad – {nombre_df} ===")
    display(acc_por_rango)

distribucion_por_rango(df_prov, "PROVINCIAS")
distribucion_por_rango(df_tot, "TOTALES")
```

PASO 11 – Histogramas de variables numéricas

```python
numeric_cols_prov = df_prov.select_dtypes(include=["int64","float64"]).columns
numeric_cols_tot  = df_tot.select_dtypes(include=["int64","float64"]).columns

print("=== Histogramas – PROVINCIAS ===")
df_prov[numeric_cols_prov].hist(bins=20, figsize=(12,6))
plt.tight_layout()
plt.show()

print("=== Histogramas – TOTALES ===")
df_tot[numeric_cols_tot].hist(bins=20, figsize=(12,6))
plt.tight_layout()
plt.show()
```

PASO 12 – Matriz de correlación (si hay más de 1 columna numérica)

```python
if len(numeric_cols_prov) > 1:
    print("=== Matriz de correlación – PROVINCIAS ===")
    plt.figure(figsize=(8,5))
    sns.heatmap(df_prov[numeric_cols_prov].corr(), annot=True, cmap="coolwarm")
    plt.title("Correlación – PROVINCIAS")
    plt.show()

if len(numeric_cols_tot) > 1:
    print("=== Matriz de correlación – TOTALES ===")
    plt.figure(figsize=(8,5))
    sns.heatmap(df_tot[numeric_cols_tot].corr(), annot=True, cmap="coolwarm")
    plt.title("Correlación – TOTALES")
    plt.show()
```

PASO 13 – Guardar archivos LIMPIOS para Excel (coma decimal + ;)

```python
def convertir_decimales_a_coma(df):
    df_export = df.copy()
    for col in df_export.columns:
        if df_export[col].dtype in ["float64", "float32"]:
            df_export[col] = df_export[col].apply(lambda x: str(x).replace('.', ','))
    return df_export

df_prov_export = convertir_decimales_a_coma(df_prov)
df_tot_export  = convertir_decimales_a_coma(df_tot)

df_prov_export.to_csv(
    "Internet_Accesos_Velocidad_Rango_Provincias_CLEAN.csv",
    sep=';',
    index=False,
    encoding='utf-8'
)

df_tot_export.to_csv(
    "Internet_Accesos_Velocidad_Rango_Totales_CLEAN.csv",
    sep=';',
    index=False,
    encoding='utf-8'
)

print("Archivos limpios guardados correctamente para Excel.")
```
