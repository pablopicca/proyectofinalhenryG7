# Vistas SQL completas (incluyendo evolución por tecnología)


1. Accesos por Tecnología (Último Trimestre)


```python

CREATE VIEW vw_Tecnologia_Ultimo AS
SELECT t.Tecnologia,
       SUM(a.Accesos) AS Total_Accesos
FROM Accesos_Ultimo_Trim a
JOIN Tecnologias t ON a.ID_Tecnologia = t.ID_Tecnologia
GROUP BY t.Tecnologia;
```

2 Vista Dinamica de Top Provincias por Accesos (Años)

```python

CREATE VIEW vw_Top_Provincias_Ultimo AS
SELECT p.Provincia,
       SUM(cp.[Total(BAF+DU)]) AS Total_Accesos
FROM Conexiones_Provincias cp
JOIN Provincias p ON cp.ID_Provincia = p.ID_Provincia
WHERE cp.ID_Periodo = '2025-T1'
GROUP BY p.Provincia
ORDER BY Total_Accesos DESC;
```

2.1 Top Provincias por Accesos (Último Trimestre)

```python


CREATE VIEW vw_Top_Provincias_Anual AS
SELECT p.Provincia,
       per.Año,
       SUM(cp.[Total(BAF+DU)]) AS Total_Accesos
FROM Conexiones_Provincias cp
JOIN Periodos per ON cp.ID_Periodo = per.ID_Periodo
JOIN Provincias p ON cp.ID_Provincia = p.ID_Provincia
GROUP BY p.Provincia, per.Año

```

3. Evolución Total de Accesos

```python

CREATE VIEW vw_Evolucion_Accesos AS
SELECT per.Año, per.Trimestre,
       SUM(cp.[Total(BAF+DU)]) AS Total_Accesos
FROM Conexiones_Periodos cp
JOIN Periodos per ON cp.ID_Periodo = per.ID_Periodo
GROUP BY per.Año, per.Trimestre
```

4. Vista Dinamica de distribucion de Rangos de velocidad por periodo.

```python

CREATE VIEW vw_Distribucion_Velocidad AS
SELECT per.ID_Periodo,
       per.Año,
       per.Trimestre,
       r.Rango_Velocidad,
       SUM(v.Accesos) AS Total_Accesos
FROM Velocidades_Prov v
JOIN Rangos r ON v.ID_Rango_Velocidad = r.ID_Rango_Velocidad
JOIN Periodos per ON v.ID_Periodo = per.ID_Periodo
GROUP BY per.ID_Periodo, per.Año, per.Trimestre, r.Rango_Velocidad
ORDER BY per.Año, per.Trimestre, Total_Accesos DESC;
```

4.1 Vista de Distribucion de Rango de Velocidad En el ultimo trimestre

```python

CREATE VIEW vw_Distribucion_Velocidad_Ultimo AS
SELECT r.Rango_Velocidad,
       SUM(v.Accesos) AS Total_Accesos
FROM Velocidades_Prov v
JOIN Rangos r ON v.ID_Rango_Velocidad = r.ID_Rango_Velocidad
WHERE v.ID_Periodo = '2025-T1'
GROUP BY r.Rango_Velocidad
ORDER BY Total_Accesos DESC;
```

5. Penetración Promedio por Provincia

```python

CREATE VIEW vw_Penetracion_Provincia AS
SELECT p.Provincia,
       AVG(cp.[Accesos cada 100 Hogares]) AS Penetracion_Hogares,
       AVG(cp.[Accesos cada 100 Habitantes]) AS Penetracion_Habitantes
FROM Conexiones_Provincias cp
JOIN Provincias p ON cp.ID_Provincia = p.ID_Provincia
GROUP BY p.Provincia;
```

6. Tecnología vs Velocidad (Último Trimestre)

```python

CREATE VIEW vw_Tecnologia_Velocidad_Ultimo AS
SELECT t.Tecnologia,
       r.Rango_Velocidad,
       SUM(a.Accesos) AS Total_Accesos
FROM Accesos_Ultimo_Trim a
JOIN Tecnologias t ON a.ID_Tecnologia = t.ID_Tecnologia
JOIN Rangos r ON a.ID_Rango_Velocidad = r.ID_Rango_Velocidad
GROUP BY t.Tecnologia, r.Rango_Velocidad;
```

7. Evolución por Tecnología (Histórico)

```python

CREATE VIEW vw_Evolucion_Tecnologia AS
SELECT per.Año, per.Trimestre, 'ADSL' AS Tecnologia, SUM(cp.ADSL) AS Accesos
FROM Conexiones_Periodos cp
JOIN Periodos per ON cp.ID_Periodo = per.ID_Periodo
GROUP BY per.Año, per.Trimestre
UNION ALL
SELECT per.Año, per.Trimestre, 'Cablemodem', SUM(cp.Cablemodem)
FROM Conexiones_Periodos cp
JOIN Periodos per ON cp.ID_Periodo = per.ID_Periodo
GROUP BY per.Año, per.Trimestre
UNION ALL
SELECT per.Año, per.Trimestre, 'Fibra Optica', SUM(cp.[Fibra Optica])
FROM Conexiones_Periodos cp
JOIN Periodos per ON cp.ID_Periodo = per.ID_Periodo
GROUP BY per.Año, per.Trimestre
UNION ALL
SELECT per.Año, per.Trimestre, 'Wireless', SUM(cp.Wireless)
FROM Conexiones_Periodos cp
JOIN Periodos per ON cp.ID_Periodo = per.ID_Periodo
GROUP BY per.Año, per.Trimestre;
```
