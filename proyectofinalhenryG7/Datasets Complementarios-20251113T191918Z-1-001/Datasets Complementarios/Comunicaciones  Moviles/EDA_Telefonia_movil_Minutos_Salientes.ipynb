## 1. Importaciones y configuración

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    'figure.figsize': (11, 5),
    'axes.titlesize': 14,
    'axes.labelsize': 12
})
```

## 2. Carga del dataset desde la ruta indicada
Se preserva la codificación **UTF‑8 con BOM** para acentos/eñes en Windows.

```python
ruta = r'C:\\Users\\u607935\\OneDrive - Telecom Argentina SA\\Escritorio\\Curso de Data Analitycs\\Proyencto Integrador Final Grupo 7\\Comunicaciones  Moviles\\Telefonía móvil Minutos Salientes_con_ID.csv'

# Cargar el archivo
df = pd.read_csv(ruta, encoding='utf-8-sig')

# Copia del crudo
df_raw = df.copy()

print('Shape:', df.shape)
print('Columnas:', list(df.columns))
display(df.head())
```

## 3. Perfil inicial del dataset
- Tipos de datos y memoria (`df.info()`).
- Valores faltantes y duplicados.
- Estadísticos descriptivos de `Total de Minutos salientes`.

```python
df.info()
print('
Nulos por columna:
', df.isna().sum())
print('Duplicados (filas completas):', df.duplicated().sum())

if 'Total de Minutos salientes' in df.columns:
    print('
Estadísticos de Total de Minutos salientes:')
    print(df['Total de Minutos salientes'].describe())
```

## 4. Validaciones de consistencia
- Conversión a numérico.
- Chequeo de negativos.
- Verificación **pospago + prepago = total minutos**.


```python
num_cols = ['Registro_ID','Año','Trimestre','Minutos pospago salientes','Minutos prepago salientes','Total de Minutos salientes','TiempoTrim_ID']
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

print('
Negativos por columna:
', (df[num_cols] < 0).sum())

# Consistencia pospago+prepago vs total
suma = df['Minutos pospago salientes'] + df['Minutos prepago salientes']
consist = (suma == df['Total de Minutos salientes'])
print(f'
Registros consistentes: {consist.mean():.2%} ({consist.sum()} de {len(consist)})')
df['dif_minutos'] = df['Total de Minutos salientes'] - suma
```

## 5. Limpieza y ordenamiento
- Eliminación de duplicados.
- Orden temporal por `Año` y `Trimestre`.
- Construcción de `Periodo` (YYYY-Qn).
- Remoción de columnas auxiliares.

```python
df = df.drop_duplicates().sort_values(['Año','Trimestre']).reset_index(drop=True)
df['Periodo'] = pd.PeriodIndex.from_fields(year=df['Año'], quarter=df['Trimestre']).astype(str)
df = df.drop(columns=['dif_minutos'], errors='ignore')
display(df.head())
```

## 6. Variables derivadas
- `porc_pospago_min` y `porc_prepago_min` (% sobre el total).
- `Total_min_QoQ_%` (variación trimestral), `Total_min_YoY_%` (t-4).
- `Total_min_MovAvg4` (media móvil 4 trimestres).

```python
df['porc_pospago_min'] = df['Minutos pospago salientes'] / df['Total de Minutos salientes'] * 100
df['porc_prepago_min'] = df['Minutos prepago salientes'] / df['Total de Minutos salientes'] * 100
df['Total_min_QoQ_%'] = df['Total de Minutos salientes'].pct_change() * 100
df['Total_min_YoY_%'] = df['Total de Minutos salientes'].pct_change(4) * 100
df['Total_min_MovAvg4'] = df['Total de Minutos salientes'].rolling(4).mean()

display(df[['Año','Trimestre','porc_pospago_min','porc_prepago_min','Total_min_QoQ_%','Total_min_YoY_%','Total_min_MovAvg4']].tail(8))
```

## 7. Visualizaciones
- Serie temporal de `Total de Minutos salientes` con media móvil.
- Participación (%) de **pospago vs prepago** (últimos 12 trimestres).

```python
import matplotlib.ticker as mtick

# 7.1 Serie temporal de minutos totales con media móvil
plt.figure()
sns.lineplot(data=df, x='Periodo', y='Total de Minutos salientes', marker='o', label='Total minutos', color='#2E86C1')
sns.lineplot(data=df, x='Periodo', y='Total_min_MovAvg4', marker='o', label='Media móvil (4)', color='#7D3C98')
plt.xticks(rotation=45, ha='right')
plt.title('Minutos salientes por trimestre (con media móvil 4)')
plt.xlabel('Periodo'); plt.ylabel('Minutos'); plt.legend(); plt.tight_layout(); plt.show()

# 7.2 Porcentaje pospago vs prepago (últimos 12 trimestres)
tail = df.tail(12)
plt.figure()
sns.lineplot(data=tail, x='Periodo', y='porc_pospago_min', marker='o', label='Pospago', color='#1ABC9C')
sns.lineplot(data=tail, x='Periodo', y='porc_prepago_min', marker='o', label='Prepago', color='#D35400')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.xticks(rotation=45, ha='right')
plt.title('Participación (%) en minutos: pospago vs prepago (últimos 12 trimestres)')
plt.xlabel('Periodo'); plt.ylabel('% del total de minutos'); plt.legend(); plt.tight_layout(); plt.show()
```

## 8. Guardado del archivo **limpio**
Se guarda el CSV en la **misma ruta** con la **leyenda `_limpio`** agregada al nombre, codificación `utf-8-sig`.

```python
ruta_path = Path(ruta)
ruta_limpia = ruta_path.with_name(ruta_path.stem + '_limpio' + ruta_path.suffix)
df.to_csv(ruta_limpia, index=False, encoding='utf-8-sig')
print('Guardado en:', ruta_limpia)
```
